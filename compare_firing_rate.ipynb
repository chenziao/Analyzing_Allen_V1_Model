```python
import pandas as pd

# Load the CSV files into pandas DataFrames
file1_path = 'data/output_movie_5s/drifting_grating_firing_rate_statisitcs.csv'
file2_path = 'data/output_movie_v_report_tau_4.0_PN2PV_10_PV2PN_10_PV2PV_0.5_BG2PN_1.5_SOM2PV_0.1/drifting_grating_firing_rate_statisitcs.csv'

df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)

common_key = 'pop_name'

# Merge the DataFrames on the common key
merged_df = pd.merge(df1, df2, on=common_key, suffixes=('_file1', '_file2'))

# Iterate over the columns you want to compare (excluding the common key)
for column in df1.columns.difference([common_key]):
    # Calculate the difference
    merged_df[f'{column}_difference'] = merged_df[f'{column}_file2'] - merged_df[f'{column}_file1']
    merged_df[f'{column}_difference_relative'] = merged_df[f'{column}_difference'] / merged_df[f'{column}_file1']

display(merged_df)
```

```python
print('layer 4 FSI has mean has a ' + str(merged_df['mean_difference'][15]) + ' difference between the runs compared')
print('layer 5 FSI has mean has a ' + str(merged_df['mean_difference'][23]) + ' difference between the runs compared')
```

```python
import matplotlib.pyplot as plt

plt.bar(merged_df['pop_name'], merged_df['mean_difference_relative'],
        color ='maroon', width = 0.4)
plt.ylabel('relative different')
plt.xticks(rotation=72, ha='center')
plt.show()
```
