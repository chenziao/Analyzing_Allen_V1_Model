```python
import h5py
from bmtk.utils.reports.compartment import CompartmentReport

output_path = '../V1_bio_outputs/output_movie_v_report_tau_4.0_PN2PV_3_PV2PN_3_PV2PV_0.5_BG2PN_5_BG2PV_5_SOM2PV_0.1/'
e4_90 = CompartmentReport(output_path + 'v-report-around90-e4.h5')
e4_270 = CompartmentReport(output_path + 'v-report-around270-e4.h5')
e5_90 = CompartmentReport(output_path + 'v-report-around90-e5.h5')
e5_270 = CompartmentReport(output_path + 'v-report-around270-e5.h5')

pv4_90 = CompartmentReport(output_path + 'v-report-around90-pv4.h5')
pv4_270 = CompartmentReport(output_path + 'v-report-around270-pv4.h5')
pv5_90 = CompartmentReport(output_path + 'v-report-around90-pv5.h5')
pv5_270 = CompartmentReport(output_path + 'v-report-around270-pv5.h5')
```

```python
import matplotlib.pyplot as plt

def plot_voltage(com_rep,plot_window=None,node_ids=None):
    if node_ids == None:
        plt.plot(com_rep.data(),label = com_rep.node_ids())
    if node_ids:
        for n in range(len(node_ids)):
            plt.plot(com_rep.data(node_id = node_ids[n]), label = node_ids[n])
    if plot_window:
        plt.xlim(plot_window)

    plt.legend()
    plt.show()

plot_voltage(e5_90)
plot_voltage(e5_270)
```
